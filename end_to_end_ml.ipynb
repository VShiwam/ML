## Data Cleaning

```python
# remove missing values
df.dropna(inplace=True) 
```

```python
# check
df.isnull().sum()
```

```python
# remove duplicates value
df.duplicated().sum()
```

```python
# drop duplicate
df.drop_duplicates(inplace=True) 
```

```python
# check
df.duplicated().sum()
```

```python
df.describe()
```

```python
# check outliers
import matplotlib.pyplot as plt

plt.boxplot(df['lead_time'],vert=False)
plt.ylabel('Variable')
plt.xlabel('lead_time')
plt.title('Box Plot')
plt.show()
```

```python
# Calculate the lower and upper bounds
Q1 =df['lead_time'].quantile(0.25)
Q3 =df['lead_time'].quantile(0.75)
IQR= Q3-Q1

# Drop the outliers
df = df[(df['lead_time'] >= (Q1 - 1.5 * IQR)) & (df['lead_time'] <= (Q3 + 1.5 * IQR))]
```

```python
# remove outliers
import matplotlib.pyplot as plt

plt.boxplot(df['lead_time'],vert=False)
plt.ylabel('Variable')
plt.xlabel('lead_time')
plt.title('Box Plot')
plt.show()
```

```python
# Data Type coversion
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
df['is_canceled'] = df['is_canceled'].astype('category')
# df['deposit_type'] = 
```
